AGRUPANDO ESTADOS POR NÚMEROS DE ÓBITOS NAS RODOVIAS

```python
mortos_uf=df.groupby(by="uf")["mortos"].size().to_frame ().reset_index ()
```

```python
x=mortos_uf.uf
y=mortos_uf.mortos
```

Quais são os estados com rodovias mais fatais no primeiro trimestre do ano de 2020 ? veja logo abaixo 👇

```python
plt.figure(figsize=(15,7))
sns.barplot(data=mortos_uf,x=x,y=y)
plt.title("Números de Mortos nas Rodovias do BR por Estado no Primeiro Trimestre de 2020",fontsize=20);
```

```python
causa_acidente = df.causa_acidente.value_counts().to_frame ().reset_index ()
```

```python
causa_acidente = causa_acidente.sort_values(by="causa_acidente")
```

■Quais são as principais causas de acidentes nas rodovias no primeiro trimestre do ano de 2020 ?

```python
causa_acidente
```

☆ Apresentação visual do resultado

```python
plt.figure(figsize=(13,6))
sns.barplot(y=causa_acidente["index"],
            x=causa_acidente["causa_acidente"])
plt.ylabel("CAUSAS DO ACIDENTE")
plt.xlabel("NÚMEROS DE ACIDENTES")
plt.title("Principais Causas de Acidentes Em Rodovias Do BR no Primeiro Trimestre de 2020",fontsize=18)
plt.show();
```

● Quais são as correlações entre as variáveis(colunas),do conjunto de dados ?

■ Quais são as condições climáticas onde mais ocorreram acidentes com vítimas fatais no primeiro trimestre do ano de 2020 nas rodovias Brasileiras ?

☆ Apresentação visual do resultado

```python
df.groupby(by="condicao_metereologica")["mortos"].sum().plot(kind="bar",color=["k"])
plt.xticks(rotation=45,ha="right")
plt.suptitle("Acidentes de Trânsito no Brasil Primeiro Trimestre do Ano de 2020",fontsize=15)
plt.title("Condições Meteorológicas em Relação aos Números de Mortos",y=0.98,fontsize=12);
```

■ Quais foram as distribuições diárias por mortes em relação aos acidentes?

```python
mortos_dia = df.groupby(by="dia_semana")["mortos"].sum().to_frame().reset_index().sort_values(by="mortos",ascending=True)
print(mortos_dia)
```

☆ Apresentação visual do resultado

```python
sns.barplot(data=mortos_dia,x="dia_semana",y="mortos")
plt.title("Números de mortos por dia da semana")
plt.suptitle("Acidentes de trânsito em rodovias Brasileiras primeiro trimestre de 2020")
plt.xticks(rotation=45,ha="right");
```

● Quais são as  distribuições de óbitos por horários em relação aos acidentes nas rodovias Brasileiras no primeiro trimestre de 2020 ?

```python
horario_mortos = df.groupby(by="horario")["mortos"].sum().to_frame().reset_index().sort_values(by="mortos")
```

```python
plt.figure(figsize=(20,10))
plt.hist(horario_mortos.horario,
         horario_mortos.mortos,color="k")
plt.xticks(rotation=45,ha="right")
plt.title("Distribuição de Óbitos por horários em relação aos acidentes nas rodovias Brasileiras no primeiro trimestre de 2020",fontsize=25);
```

É perceptível os registros de óbitos nas barras mais elevados do histograma chegando a apresentar cerca de quinhentas mortes entre os horários 13:57:00, 13:59:00, 22:35:00, 14:01:00, 14:03:00.

■ Quais as distribuições de óbitos por BRs ?

```python
br_mortos = df.groupby(by="br")["mortos"].sum().to_frame().reset_index().sort_values(by="mortos")
```

```python
br_mortos = br_mortos[br_mortos["mortos"]>0]
```

☆ Apresentação visual do resultado

```python
plt.figure(figsize=(20,8))
sns.barplot(data=br_mortos,x="br",y="mortos")
plt.xticks(rotation=90,alpha=0.7)
plt.title("Distribuições de óbitos por BRs em relação aos acidentes de trânsito primeiro trimestre de 2020",fontsize=25);
```

!Observação para as BRS 101 e 116 contendo números elevados de mortos em acidentes no primeiro trimestre do ano de 2020
