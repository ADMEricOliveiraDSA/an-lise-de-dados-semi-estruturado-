```python
# EXECUTANDO IMPORTES 

import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 

%matplotlib inline 
```

```python
# LENDO ARQUIVO CSV 
# Pós graduação de 2013 a 2020 
df = pd.read_csv ('posgraduacao_lato_2019.csv',sep=';',encoding='cp1252')

df.head()
```

```python
# DESCOBRINDO VALORES ESTATÍSTICOS 
df.describe()
```

```python
#VENDO INFORMAÇÕES SOBRE TIPOS DO DF
df.info ()
```

```python
# PLOTANDO GRÁFICO POR N° DE SETORES DE CURSOS OFERTADOS DE 2013 A 2020 
df['Nome_Setor'].value_counts().plot.bar ();
plt.title('Número de cursos por setores de 2013 a 2020 em graduação e pós-graduação ')
plt.show();
```

```python
#CRIANDO VARIÁVEIS PARA PLOT 
x = df ['Curso_ConcEspec']
y = df ['Curso_ConcAperf']
```

```python
# OBSERVANDO A RELAÇÃO EM NÚMEROS DE CURSOS DE ESPECIALIZAÇÃO E APERFEIÇOAMENTO 
# 50 dos cursos dentre os 14 setores são os mais oferecidos em especialização
plt.plot(x, y, 'bo')
plt.show();
```

```python
data = df ['Ano'].value_counts()
plt.plot(data)
plt.title('ANOS COM MAIS E MENOS OFERTAS DE CURSOS NOVOS NAS GRADES')
plt.grid(color='r')
plt.ylim (40,150)
plt.show();
# Entre 2013 e 2020, 2018 foi o ano onde foram lançados mais cursos nas grades de graduação e pós-graduação chegando a 130.
```

```python
# HISTOGRAMA DE RELAÇÃO AOS CURSOS POR ANO
# NÚMEROS DE CURSOS MATRÍCULADOS PARA CADA ANO EM TAL INSTITUIÇÃO 
df [['Ano','Curso_Matriculados']].hist();
```

```python
fig = df.plot (subplots=True)
plt.figure(figsize=(20,10))

plt.show();
# Análise visual de picos e oscilações entre relações de cursos novos.  
# Dentre os anos, cursos matriculados,cursos com especialização e com aperfeiçoamento.
```
