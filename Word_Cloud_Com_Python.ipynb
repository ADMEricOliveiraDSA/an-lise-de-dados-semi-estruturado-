#### Criando nuvens de palavras contendo as palavras com sentidos que mais me interessam e que tem relação com minha vivência.

```python
# Importando bibliotecas para criação das nuvens de palavras 
import numpy as np 
import matplotlib.pyplot as plt
from wordcloud import WordCloud
%matplotlib inline
```

```python
# Definindo as palavras ao texto quê seram exibidos na wordcloud 
# Quanto mais uma palavra se repete maior será seu tamanho na nuvem de palavras
# Por ser extremamente cristão, usarei a palavra Deus mais vezes, pois pra mim ele está acima de todas as coisas
 
text = ("Eric batalhas provações estudos família saúde Eric Educação Deus batalhas Deus administração  Deus natureza Deus administração família Eric Data Science Deus ciência dados análise  dados Deus matemática Eric estatística Deus machine learning família Python inglês Python")
```

```python
# Criando a imagem a ser exibida
# Criando variavel que recebe os atributos de tamanho de imagem e geração do texto 
# Excluindo a margem para não mostrar os números equivalentes a resolução da imagem 
 
wordcloud = WordCloud(width=1280,height=640, margin=0).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
 
plt.show()
```

```python
# Mesmo processo da imagem anterior ,porém trocando o mapeamento de cores para uma fonte azul.
 
wordcloud = WordCloud(width=1280, height=640, colormap="Blues").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

```python
# Mesmo processo da imagem anterior ,porém trocando o mapeamento de cores para uma fonte céu azul.
 
wordcloud = WordCloud(width=1280, height=640, background_color="skyblue").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

Existem inúmeras possibilidades de produção de WordCloud, diferentes escopos,formas ,fontes, aguarde mais novidades.
