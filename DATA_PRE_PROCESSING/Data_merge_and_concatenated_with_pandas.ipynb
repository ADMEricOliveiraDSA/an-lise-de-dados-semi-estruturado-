# Leitura das Tabelas 

```python
# Importes 
import pandas as pd 
# Carregando a base de testes 
df1 = pd.read_excel('Excel_Produtos.xlsx', sheet_name=0,engine='openpyxl')
df2 = pd.read_excel('Excel_Produtos.xlsx', sheet_name=1,engine='openpyxl')
```

```python
df1 = df1.head(10)
df1
```

```python
df2 = df2.head(10)
df2
```

# Mesclagem dos dados 

## A biblioteca Pandas em python fornece uma única função, merge, como o ponto de entrada para todas as operações de junção de banco de dados padrão entre objetos DataFrame

```python
# Junção completa dos dois DataFrames com ordem dos índices.
pd.merge(df1, df2,
         how='outer',      # similar full-join no SQL / Unir completamente 
         left_index=False, # usa índice esquerdo como chave de junção 
         right_index=False,# usa índice direto como chave de junção 
         sort=True)        #ordem das chaves de junção obedecendo o índice 
```

```python
# Unindo os DataFrames lado a lado 
pd.merge(df1,df2, 
         how='outer',      
         left_index=True,  # Confirma união pela esquerda 
         right_index=True, # Confirma união pela direita 
         sort=True)        
```

# Concatenação dos dados 

## O Pandas oferece vários recursos para combinar facilmente objetos Series, DataFrame e Painel.  No exemplo a seguir, a função concat executa operações de concatenação ao longo de um eixo.  

```python
# União padrão, índices não são transformados!  
pd.concat([df1,df2])
```

```python
# União lateral, completa pelo eixo 1 (linhas)
pd.concat([df1,df2],axis=1,join='outer')
```
