# P-VALUE  ( VALOR DE P )

### O valor p é sobre a força de uma hipótese.  Construímos hipóteses com base em algum modelo estatístico e comparamos a validade do modelo usando o valor de p.  Uma maneira de obter o valor p é usando o teste t.

### Na maioria das análises, um alfa de 0.05 é usado como ponto de corte para significância. Se o valor-p for menor que 0.05, devemos rejeitar a hipótese nula de que não há diferença entre as médias e concluir que existe uma diferença significativa. Se o valor-p for maior que 0.05, não é possível concluir que existe uma diferença significativa.

### Nos exemplos a seguir, há duas amostras, que podem vir da mesma distribuição ou de distribuições diferentes, e queremos testar se essas amostras têm as mesmas propriedades estatísticas.

### ttest_ind - Calcula o teste T para as médias de duas amostras independentes de pontuação.  Este é um teste bilateral para a hipótese nula de que duas amostras independentes têm valores médios (esperados) idênticos.  Este teste assume que as populações têm variações idênticas por padrão.

### Podemos usar este teste, se observarmos duas amostras independentes da mesma população ou de populações diferentes.  Vamos considerar o seguinte exemplo.

# Abaixo de 0.05, significativo. Acima de 0.05, não significativo.

##  H0 ( p-value < 0.05 ) - Rejeita-se hipótese nula 
##  H1 ( p-value > 0.05 ) - Aceita-se hipótese alternativa 

```python
# PACOTE ESTATÍSTICO 
from scipy import stats
# VARIÁVEIS ALEATÓRIAS, LOCALIZAÇÃO, ESCALA E TAMANHO
rvs1 = stats.norm.rvs(loc = 5,scale = 10,size = 500)
rvs2 = stats.norm.rvs(loc = 5,scale = 10,size = 500)
# EXIBIR RESULTADO DO TESTE 
print(stats.ttest_ind(rvs1,rvs2))
```

# RESPOSTA = Os resultados não foram estatisticamente significativos para aceitar quê há uma diferença abrangente entre as médias encontradas nos dados da amostra.

# QUI-QUADRADO ( Chi-Square Test )

### O teste Qui-quadrado é um método estatístico para determinar se duas variáveis ​​categóricas têm uma correlação significativa entre elas.  Ambas as variáveis ​​devem ser da mesma população e devem ser categóricas como - Sim / Não, Masculino / Feminino, Vermelho / Verde etc. Por exemplo, podemos construir um conjunto de dados com observações sobre o padrão de compra de sorvete das pessoas e tentar correlacionar o gênero de uma pessoa com o sabor do sorvete de sua preferência.  Se uma correlação for encontrada, podemos planejar um estoque apropriado de sabores, sabendo o número de gênero das pessoas que os visitam.

```python
# Importações de pacotes 
from scipy        import stats
import numpy             as np
import matplotlib.pyplot as plt
%matplotlib inline 

# Base, matriz 2D de 100 valores variando de 0 a 10
x = np.linspace(0, 10, 100)

# Estilo de linhas, grau de liberdade, tamanho da figura  
linestyles = [':', '--', '-.', '-']
grau_de_liberdade = [1, 4, 7, 6]
fig,ax = plt.subplots(1,1,figsize = (20,10))

# Para cada dado na base, junta-os e aplica o teste. Depôs adiciona ao plot e cria a figura.
# Função de densidade de probabilidade avaliada em (X) + Grau de liberdade 
for df, ls in zip(grau_de_liberdade, linestyles):
    ax.plot(x, stats.chi2.pdf(x, df), linestyle=ls, linewidth=5.0 )
    
# Limite de índice da linha de cada eixo 
plt.xlim(0, 10)
plt.ylim(0, 0.4)

# Rótulos e título 
plt.xlabel('Valor',fontsize = 35)
plt.ylabel('Frequencia',fontsize = 35)
plt.title('Qui-quadrado - ( Chi-Square Test )',fontsize = 40)

# Exibir  
plt.show()
```
